```python
import matplotlib.pyplot as plt
import torch

def draw_acc_history(histories):
    
    plt.figure()
    for act, history in histories:
        plt.plot(torch.tensor(history["acc"]), label = f'{act}_train')
        plt.plot(torch.tensor(history["test_acc"]), label = f'{act}_test')
    plt.xlabel('Epoch')
    plt.ylabel('acc')
    plt.legend()

def draw_loss_history(histories):
    
    plt.figure()
    for act, history in histories:
        plt.plot(torch.tensor(history["loss"]), label = f'{act}_train')
        plt.plot(torch.tensor(history["test_loss"]), label = f'{act}_test')
    plt.xlabel('Epoch')
    plt.ylabel('loss')
    plt.legend()

```
